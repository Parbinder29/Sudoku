```python
class Solution(object):  # Define a class named Solution

    def solveSudoku(self, board):  # Main function to solve Sudoku

        # Create sets to store used numbers in each row, column, and box
        rows = [set() for _ in range(9)]   # rows[i] stores numbers in row i
        cols = [set() for _ in range(9)]   # cols[i] stores numbers in column i
        boxes = [set() for _ in range(9)]  # boxes[i] stores numbers in 3x3 box i

        # List to store positions of empty cells
        empty = []

        # Step 1: Initialize sets and empty cells
        for r in range(9):  # Loop through each row
            for c in range(9):  # Loop through each column

                if board[r][c] == '.':  # If cell is empty
                    empty.append((r, c))  # Store its position

                else:
                    num = board[r][c]  # Get the number in the cell

                    rows[r].add(num)  # Add number to corresponding row set
                    cols[c].add(num)  # Add number to corresponding column set

                    # Compute box index and store number
                    boxes[(r // 3) * 3 + (c // 3)].add(num)
                    # (r//3)*3 + (c//3) gives index from 0 to 8 for 3x3 boxes

        def solve(i):  # Recursive function to fill empty cells

            # Base case: if all empty cells are filled
            if i == len(empty):
                return True  # Sudoku solved

            r, c = empty[i]  # Get current empty cell position

            # Find corresponding 3x3 box index
            box_index = (r // 3) * 3 + (c // 3)

            # Try numbers from 1 to 9
            for num in '123456789':

                # Check if number is not already used
                if num not in rows[r] and num not in cols[c] and num not in boxes[box_index]:

                    # Place the number in the board
                    board[r][c] = num

                    # Update sets
                    rows[r].add(num)
                    cols[c].add(num)
                    boxes[box_index].add(num)

                    # Recur for next empty cell
                    if solve(i + 1):
                        return True  # If solved, return True

                    # Backtracking step (undo changes)
                    board[r][c] = '.'  # Remove number from board

                    rows[r].remove(num)  # Remove from row set
                    cols[c].remove(num)  # Remove from column set
                    boxes[box_index].remove(num)  # Remove from box set

            return False  # No valid number found → backtrack

        solve(0)  # Start solving from first empty cell
```
