```python
class Solution(object):
    def isValidSudoku(self, board):
        res = []  # List to store row, column, and box representations

        for i in range(9):  # Loop through rows (0 to 8)
            for j in range(9):  # Loop through columns (0 to 8)
                element = board[i][j]  # Get current cell value

                if element != '.':  # Ignore empty cells
                    # Add three types of entries:

                    # 1. Row check → (row index, number)
                    #    ensures number is unique in its row
                    # 2. Column check → (number, column index)
                    #    ensures number is unique in its column
                    # 3. Box check → (box row, box column, number)
                    #    ensures number is unique in its 3x3 sub-box

                    res += [
                        (i, element),                 # Row representation
                        (element, j),                 # Column representation
                        (i // 3, j // 3, element)     # 3x3 box representation
                    ]

        # Convert list to set to remove duplicates
        # If duplicates exist → lengths differ → invalid Sudoku
        return len(res) == len(set(res))
```
